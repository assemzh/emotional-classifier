```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
def formatWordUrl(inputWord):
    url = 'http://www.thesaurus.com/browse/'
    url = url + inputWord.strip().lower().replace(' ', '%20')
    return url
```

```python
def fetchWordData(inputWord):
    url = formatWordUrl(inputWord)
    r = requests.get(url)
    
    soup = BeautifulSoup(r.content, 'html.parser')
    
    div = soup.find("div", { "id" : "meanings" })
    
    syn = [ x.find("a").text[:-1] for x in div.find_all("li") if x.find("a") is not None]
    return syn
```

```python
happiness = fetchWordData('sadness')

for i in fetchWordData('sadness'):
    temp = fetchWordData(i)
    happiness.extend(temp)
    
print(len(happiness))
```

```python
happiness = list(set(happiness))
print(len(happiness))
```
